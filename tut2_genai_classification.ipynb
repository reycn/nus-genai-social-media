# Topic Modeling and Sentiment Analysis Using the OpenAI API

```python
import os
from openai import OpenAI
from rich import print as pp
```

# 1. Prepare the Environment

## 1.1. Initialize the OpenAI Client

```python
# Try reading the environment variable
try:
    OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
    if OPENAI_API_KEY is None:
        raise KeyError
    pp("[green]Environment variable OPENAI_API_KEY was read successfully.[/green]")
except KeyError:
    OPENAI_API_KEY = None
    pp("[red]Environment variable OPENAI_API_KEY was not found. Please make sure it is set correctly.[/red]")

client = OpenAI(api_key=OPENAI_API_KEY)  # Replace with your API key
```

# 1. Sentiment Classification


## 1.1. Define the Method

```python
def analyze_sentiment(text: str) -> str:
    """
    Perform sentiment analysis on the given text using the GPT-4 model.

    Parameters:
        text (str): The text to analyze for sentiment.

    Returns:
        str: The sentiment analysis result returned by GPT-4.
    """
    # Build the sentiment analysis prompt
    system_prompt = "You are a professional sentiment analysis assistant who can accurately identify and explain sentiment in text. Return `1` for positive, `-1` for negative, and `0` for neutral. If uncertain, provide the best possible guess. Return only the result."
    user_prompt = f"Please perform sentiment analysis on the following text:\n\n\"{text}\"\n\nSentiment analysis result:"

    # Call the GPT-4 model
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )

    # Extract and return the result
    return response.choices[0].message.content.strip()

result_map = {
    "1": "Positive",
    "-1": "Negative",
    "0": "Neutral"
}
```

## 1.2 Run

### Example: Positive

```python
# Provide a sample text for sentiment analysis
sample_text = "I feel very happy today because I received an unexpected surprise!"
# Call the sentiment analysis function
result = analyze_sentiment(sample_text)
# Print the result
pp(f"[blue]Text:[/blue] {sample_text}\n[green]Sentiment:[/green] {result_map.get(result, 'Unknown')}")
```

### Example: Negative

```python
# Provide a sample text for sentiment analysis
sample_text = "I had terrible luck today and couldn't get tickets for the concert I wanted."
# Call the sentiment analysis function
result = analyze_sentiment(sample_text)
# Print the result
pp(f"[blue]Text:[/blue] {sample_text}\n[green]Sentiment:[/green] {result_map.get(result, 'Unknown')}")
```

# 2. Topic Modeling


## 2.1. Define the Method

```python
def topic_modeling(text: str) -> str:
    """
    Perform topic modeling on the given text using the GPT-4 model.

    Parameters:
        text (str): The text to analyze for topics.

    Returns:
        str: The topic returned by GPT-4.
    """
    # Build the topic modeling prompt
    system_prompt = "You are a professional topic modeling assistant who can accurately identify the main topic in text. Analyze the text and return the single most relevant topic from ['Politics', 'Entertainment']. If uncertain, provide the most reasonable guess. Return only the result."
    user_prompt = f"Please perform topic modeling on the following text:\n\n\"{text}\"\n\nTopic:"

    # Call the GPT-4 model
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )

    # Extract and return the result
    return response.choices[0].message.content.strip()
```

## 2.2. Run

### Example: Politics

```python
topic_modeling("China's Ministry of National Defense held a press conference")
```

### Example: Entertainment

```python
topic_modeling("Taylor Swift opened a TikTok account and attracted young audiences")
```
