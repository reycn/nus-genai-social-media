```python
# Small script: count how many people are in an image using YOLO + Hugging Face cache
from ultralytics import YOLO
from huggingface_hub import hf_hub_download
from huggingface_hub.utils import LocalEntryNotFoundError

# YOLO weights hosted on Hugging Face
repo_id = "Ultralytics/YOLO26"
filename = "yolo26n.pt"

# 1) Try local cache only
# 2) If not found, auto-download from Hugging Face
try:
    model_path = hf_hub_download(
        repo_id=repo_id,
        filename=filename,
        local_files_only=True,
    )
    print(f"Using cached model: {model_path}")
except LocalEntryNotFoundError:
    print("No local cache found. Downloading model from Hugging Face...")
    model_path = hf_hub_download(
        repo_id=repo_id,
        filename=filename,
        local_files_only=False,
    )

model = YOLO(model_path)

# Update this path to your image
image_path = "./example/father.jpeg"

# Run inference
results = model(image_path, conf=0.25)

# COCO class id for person is 0
person_count = 0
for r in results:
    if r.boxes is None:
        continue
    cls = r.boxes.cls.int().tolist()
    person_count += sum(1 for c in cls if c == 0)

print(f"People detected: {person_count}")

# Optional: save a visualization with bounding boxes
results[0].save(filename="./example/father.extracted.jpeg")
```
